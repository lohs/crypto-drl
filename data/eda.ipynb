```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from pandas.plotting import table

# btc_min_5_df = pd.read_csv('BTC-USD_min_5.csv', sep=',', index_col=0)
# btc_min_5_df.index = pd.to_datetime(btc_min_5_df.index, unit='s')
# btc_min_15_df = pd.read_csv('BTC-USD_min_15.csv', sep=',', index_col=0)
# btc_min_15_df.index = pd.to_datetime(btc_min_15_df.index, unit='s')
# btc_hr_1_df = pd.read_csv('BTC-USD_hr_1.csv', sep=',',index_col=0)
# btc_hr_1_df.index = pd.to_datetime(btc_hr_1_df.index, unit='s')
btc_hr_24_df = pd.read_csv('BTC-USD_hr_24.csv', sep=',',index_col=0)
btc_hr_24_df.index = pd.to_datetime(btc_hr_24_df.index, unit='s')
```

```python
pairs = ["BTC-USD", "ETH-USD"]
time_periods = ["min_5", "min_15", "hr_1", "hr_24"]

for pair in pairs:
    for period in time_periods:
        # df read csv
        df = pd.read_csv(f'{pair}_{period}.csv', sep=',',index_col=0)
        df.index = pd.to_datetime(df.index, unit='s')
        df['close'] = df['close'].pct_change()
        desc = df['close'].describe()

        plot = plt.subplot(111, frame_on=False)

        #remove axis
        plot.xaxis.set_visible(False) 
        plot.yaxis.set_visible(False) 

        #create the table plot and position it in the upper left corner
        table(plot, desc,loc='upper right')

        #save the plot as a png file
        plt.savefig(f'./figures/{pair}_{period}_close.png')
        #print(f'{pair}_{period}_close.png')
```
